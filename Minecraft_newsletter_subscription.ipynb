The data set is composed of two files: players.csv includes one observation for each unique player along with demographic information and whether the player subscribed to the newsletter. sessions.csv includes one observation for each play session, including behavioral information such as session length and timing.

**Section 1**

players.csv

Each row represents a unique player.

* experience: categorical ordinal — player's self-reported, or system-classified, skill level (e.g., Pro, Veteran, Regular, Amateur, Beginner).

* subscribe: logical / binary — whether the player subscribed to the newsletter (True or False). This is the response (target) variable for your project.

* hashedEmail: string identifier — the hashed version of the player’s email used to link with sessions.csv. No personal information, but repeated values indicate multiple sessions for that player.

* played_hours: numeric — Total number of hours the player has spent on the server. May include outliers or zeros (for inactive players).

* name: categorical — player’s in-game display name. Some names may repeat or have inconsistent capitalization.

* gender: categorical — player’s reported gender. May include missing or inconsistent entries.

* Age: numeric — player’s age in years. May contain missing or unrealistic values (too young or too old).

sessions.csv

Each row represents one play session, associated with players through hashedEmail.

* hashedEmail: string identifier. links sessions to the player in players.csv.

* start_time: date/time (string). start date and time of the session; will need to be parsed properly

* end_time: date/time (string). End date and time of the session; should match the same format as start_time.

* original_start_time: numeric timestamp. Original UNIX-style timestamp for the start time; redundant but useful for precise time calculations.

* original_end_time: numeric timestamp. Original UNIX-style timestamp for the end time.

Session duration is a derived variable that can be computed, for later use, by subtracting start_time from end_time.

The data set players.csv contains one record per player, with demographic and engagement information such as skill level (experience), gender, and total hours played. The target variable to predict is subscribe, which denotes the newsletter subscription status.
The sessions.csv dataset contains recordings of game-play sessions for each player, joined by a common hashedEmail. Each session has both a human-readable and a numeric timestamp for start and end times. These will enable session duration calculations and the identification of activity patterns later on.
Potential issues to consider are inconsistent time formats, possible missing demographic values, and outliers in playtime.

**Section 3**

Before performing any predictive analysis, its best to understand the variables in the players.csv dataset and their relationship since the goal is to look at the "experience" explanatory variable and see how it relates to the response variable which in this case is the "subscribe" variable. This could show whether a player would subscribe to the newsletter or not based on their level of play and how invested they might be to the community and news of the game.

I will start by summarizing numeric variables to understand the general characteristics of the players, such as average age and total hours played. Then, I'll create visualizations to observe how newsletter subscription patterns vary across experience levels. It helps in understanding possible trends and issues to be considered in later modeling stages, like category sizes and missing values.

```r
players|>
summarise(mean_age=round(mean(Age, na.rm=TRUE), 2),
          mean_played_hours=round(mean(played_hours, na.rm=TRUE), 2))
```

```r
players |>
  count(experience, name = "n") |>
  ggplot(aes(x = experience, y = n)) +
  geom_col() +
  labs(
    title = "Number of Players by Experience Level",
    x = "Experience level",
    y = "Count of players"
  ) +
  theme_minimal()
```

This plot here displays the number of players that fall into each level of experience. The Amateur and Veteran players are the majority and represent the larger group of the entire playerbase and the pro players being the minority. Pro level players are considered to be the smallest group of players. Uneven category sizes show us that experience levels are not evenly represented, which should be considered later when evaluating relationships or building predictive models.

```r
players |>
  group_by(experience) |>
  summarise(subscribers = sum(subscribe, na.rm = TRUE),
    total = n(),
    prop_subscribed = subscribers / total) |>
  ggplot(aes(x = experience, y = prop_subscribed)) +
  geom_col() +
  labs(title = "Newsletter Subscription Rate by Experience Level",
    x = "Experience level",
    y = "Proportion subscribed (0–1)") +
  theme_minimal()
```

> Differences between groups are modest; Beginner and Regular players show slightly higher subscription rates, but the variation is small overall.

This plot compares the proportion of subscribers for each experience level: Regular and Beginner have the highest subscription rates, at approximately 0.75–0.8, while Amateur, Veteran, and Pro are a little lower. While this may indicate that the newer or mid-level experienced players are somewhat more engaged with the newsletter, the differences are relatively small overall. What the pattern really suggests is a weak positive relationship between experience level and subscription rate, rather than a strong and clear trend.

```r
ggplot(players, aes(x = experience, fill = subscribe)) +
  geom_bar(position = "fill") +
  labs(
    title = "Subscription Composition within Each Experience Level",
      subtitle = "Proportions of subscribed vs. non-subscribed players",
    x = "Experience level",
    y = "Proportion within experience group",
    fill = "Subscribed"
  ) +
  theme_minimal()
```

> Differences between groups are modest; Beginner and Regular players show slightly higher subscription rates, but the variation is small overall.

The stacked bar chart depicts the share of subscribers, in blue, and non-subscribers, in red, across all experience groups. All groups are relatively similar in their mix of subscribed and unsubscribed players, though Regular and Beginner categories have a modestly higher share of subscribers. This is supportive of a mild association between experience and subscription, but also confirms that it is not a large effect.

Overall, this exploratory analysis shows only slight differences in subscription behavior across experience levels. Regular and Beginner players appear somewhat more likely to subscribe, but the variation across groups is small. This would suggest that experience alone is not a strong predictor of newsletter subscription, though it may still contribute modestly when combined with other variables such as playtime or age in later analyses.

**Section 4**

To address the question “Can a player’s experience level predict whether they subscribe to the newsletter?”, I plan to use a logistic regression model. This method is suitable for predicting a binary response variable, such as subscribe, using one or more explanatory variables. Logistic regression is interpretable, widely used in introductory data science, and appropriate for the scope of this project.

Logistic regression models the probability of an event, in this case, a player subscribing to the newsletter, conditioned on one or more predictors. The technique can accommodate categorical predictors like experience, which will be translated into binary indicator variables for each level of experience. This way, comparisons between groups of players can be assessed while maintaining the model simple and interpretable.

Assumptions needed:

* The observations are independent, each player’s decision concerning subscription is independent of others.

* The log-odds of the response variable, subscribe, is linearly related to the predictors.

* No strong multicollinearity between predictors (which is not an issue here since only one predictor is used).

* Sufficient sample size for each category of experience.

Potential limitations:

* Logistic regression assumes a simple linear relationship on the log-odds scale and may not capture complex nonlinear behavior.

* Uneven category sizes in experience might make the estimates less certain for smaller groups.

* Because the exploratory analysis presented only weak differences between levels of experience, the predictive performance of the model is also limited.


Data processing and evaluation plan:

* The data should be split into 80% training and 20% testing sets before modeling in order to evaluate performance on unseen data.

* I would make sure that the split is stratified and maintains similar proportions of subscribers across both sets.

* Accuracy, ROC-AUC, and confusion matrix metrics would then be used to evaluate the model's performance.

* If time permitted, I would perform cross-validation on the training set to check the model's stability.

* Future work may involve adding variables like played_hours or aggregated session statistics, further improving performance.
